```python
import pandas as pd
from pathlib import Path

raw_path = Path("../data/raw")
csv_files = sorted(raw_path.glob("*.csv"))

csv_files
```

```python
summary = []

for file in csv_files:
    df = pd.read_csv(file)
    
    summary.append({
        "table_name": file.name,
        "rows": df.shape[0],
        "columns": df.shape[1],
        "duplicate_rows": df.duplicated().sum(),
        "missing_values_total": df.isna().sum().sum()
    })

inventory_df = pd.DataFrame(summary)
inventory_df
```

```python
for file in csv_files:
    df = pd.read_csv(file)
    print(f"\n=== {file.name} ===")
    display(df.isna().sum().sort_values(ascending=False).head(15))
```

```python
orders = pd.read_csv(raw_path / "orders.csv")
reviews = pd.read_csv(raw_path / "order_reviews.csv")

orders_set = set(orders["order_id"])
reviews_set = set(reviews["order_id"])

print("Orders tanpa review:", len(orders_set - reviews_set))
print("Reviews tanpa order:", len(reviews_set - orders_set))
print("Order dengan lebih dari 1 review:", (reviews["order_id"].value_counts() > 1).sum())
```

```python
date_columns = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_columns:
    orders[col] = pd.to_datetime(orders[col], errors="coerce")

orders[date_columns].isna().sum()
```
